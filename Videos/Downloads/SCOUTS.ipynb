```python
# Function to calculate AI Rating based on the formula
def calculate_ai_rating(row):
    experience_weight = 0.3
    signings_weight = 0.4
    verified_weight = 2.0

    verified_status = 1 if row['verified_status'] else 0  # Convert boolean to integer
    
    ai_rating = (row['experience_years'] * experience_weight) + \
                (row['successful_signings'] * signings_weight) + \
                (verified_status * verified_weight)
    
    return ai_rating

# Apply the function to the DataFrame
df['AI_rating'] = df.apply(calculate_ai_rating, axis=1)
```
